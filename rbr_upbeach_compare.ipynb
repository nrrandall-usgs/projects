Plot data from 3 poles closest to water: Upper Beach, Pole 0, Pole 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import datetime
```

```python
# path to corrected pressure data
pname = 'C:/Users/nrrandall/NRR/DUNEX/press_corr/'
# rbrs
fn08 = 'rbr208008_corr.cdf'
fn09 = 'rbr208009_corr.cdf'
fn10 = 'rbr208010_corr.cdf'
```

```python
ds08 = xr.open_dataset(pname + fn08)
ds09 = xr.open_dataset(pname + fn09)
ds10 = xr.open_dataset(pname + fn10)
```

```python
dt08 = ds08.dt.values
p08 = ds08.dbar.values
dt09 = ds09.dt.values
p09 = ds09.dbar.values
dt10 = ds10.dt.values
p10 = ds10.dbar.values
```

```python
# show first and last time steps of the timeseries
print([dt08[0],dt08[-1]])
print([dt09[0],dt09[-1]])
print([dt10[0],dt10[-1]])
```

```python
plt.plot(dt08, p08, label='Pole 1', zorder=2)
plt.plot(dt09, p09, label='Pole 0', zorder=1)
plt.plot(dt10, p10, label='Upper Beach Pole', zorder=0)

plt.xticks(rotation=45)
plt.ylabel('Corrected Atmospheric Pressure (dbar)')
plt.legend()
plt.show()
```

```python
plt.plot(dt08, p08, '.', label='Pole 1', zorder=2)
plt.plot(dt09, p09, '.', label='Pole 0', zorder=1)
# plt.plot(dt10, p10, '.', label='10', zorder=0)

dstart = datetime.date(2021,10,8)
dend = datetime.date(2021,10,12)

plt.xlim([dstart, dend])
plt.xticks(rotation=45)
plt.ylabel('Corrected Atmospheric Pressure (dbar)')
plt.legend()
plt.show()
```
